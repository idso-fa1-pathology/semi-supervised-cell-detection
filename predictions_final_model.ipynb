```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def read_points_and_labels_from_file(file_path):
    points_with_labels = set()
    with open(file_path, 'r') as file:
        lines = file.readlines()
    for line in lines:
        values = line.strip().split(' ')
        if len(values) == 5:
            label, x_center, y_center, _, _ = map(float, values)
            points_with_labels.add((label, x_center, y_center))
    return points_with_labels

def darken_image(image, factor=0.7):
    # Ensure factor is between 0 and 1
    factor = max(min(factor, 1), 0)
    # Convert to float to avoid truncation issues during multiplication
    image = np.array(image, dtype=np.float64)
    image = image * factor
    # Clip values to be between 0 and 255 and convert back to uint8
    image = np.clip(image, 0, 255).astype(np.uint8)
    return image


image_path = f'/Users/yshokrollahi/Desktop/MD/Project3/dataset/IF/ROIs/Validation_By_Pathologist/Karina/{name}/images/{name}_PATCH_{patch}.png'
image = cv2.imread(image_path)
height, width, channels = image.shape
dpi = 600
figsize = width / float(dpi), height / float(dpi)

add_file_path = f"/Users/yshokrollahi/Desktop/MD/Project3/dataset/IF/ROIs/Validation_By_Pathologist/Karina/{name}/predicted_annotaions/{name}_PATCH_{patch}_add.txt"
new_file_path = f"/Users/yshokrollahi/Desktop/MD/Project3/dataset/IF/ROIs/Validation_By_Pathologist/Karina/{name}/predicted_annotaions/{name}_PATCH_{patch}_new.txt"

add_points_labels = read_points_and_labels_from_file(add_file_path)
new_points_labels = read_points_and_labels_from_file(new_file_path)

additional_points_labels = new_points_labels - add_points_labels

colors = {
    0: (0, 0, 255),     # Blue
    1: (255, 255, 0),   # Cyan
    2: (255, 0, 255),   # Magenta
}

image_width, image_height = 640, 640

# Plot 'add' points (old points)
for label, x_center, y_center in add_points_labels:
    cv2.circle(image, (int(x_center * image_width), int(y_center * image_height)), radius=7, color=colors[int(label)], thickness=-1)

# Darken the entire image, including the old points
image = darken_image(image, factor=0.4)

# Plot new points with the same class colors but with distinct highlighting
for label, x_center, y_center in additional_points_labels:
    point_position = (int(x_center * image_width), int(y_center * image_height))
    # Draw the point with original class color
    cv2.circle(image, point_position, radius=7, color=colors[int(label)], thickness=-1)  
    # Optionally, draw an additional circle or feature to highlight the point

# Convert the image from BGR to RGB and display/save it as before
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=figsize, dpi=dpi)
plt.imshow(image_rgb)
plt.axis('off')
plt.tight_layout(pad=0)

save_path = f'/Users/yshokrollahi/Desktop/MD/Project3/dataset/IF/ROIs/Validation_By_Pathologist/Karina/{name}/prediction/{name}_PATCH_{patch}_prediction_difference.png'
plt.savefig(save_path, bbox_inches='tight', pad_inches=0, dpi=dpi)
```

# Batch Prediction

```python
import os
from ultralytics import YOLO
import torch
import cv2



# Load a pretrained YOLOv8 model
model = YOLO('/Users/yshokrollahi/Desktop/MD/Project3/dataset/IF/ROIs/full_PA/all_aug/predictions/detect/train/weights/best.pt')  # or the path to your custom model


# Set the minimum confidence threshold
conf_threshold = 0.001
model.conf = conf_threshold  # Set the confidence threshold for the model

# Directory containing the images
image_folder = '/Users/yshokrollahi/Desktop/MD/Project3/dataset/IF/ROIs/full_PA/all_aug/predictions/batch_images'

# Directory to save the results
output_folder = '/Users/yshokrollahi/Desktop/MD/Project3/dataset/IF/ROIs/full_PA/all_aug/predictions/batch_prediction'
os.makedirs(output_folder, exist_ok=True)

# List all images in the folder
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Process each image
for image_file in image_files:
    image_path = os.path.join(image_folder, image_file)
    results = model(image_path)

    # Assuming there's only one result per image, access the first element
    result = results[0]

    # Save the results in a text file
    txt_path = os.path.join(output_folder, image_file.split('.')[0] + '.txt')
    with open(txt_path, 'w') as f:
        for box in result.boxes:
            # Extract and format class ID and confidence
            class_id = int(box.cls) if isinstance(box.cls, torch.Tensor) else box.cls
            conf = float(box.conf) if isinstance(box.conf, torch.Tensor) else box.conf

            # Skip detections below the confidence threshold
            if conf < conf_threshold:
                continue

            # Extract and format bounding box coordinates
            bbox = box.xywh.cpu().numpy() if isinstance(box.xywh, torch.Tensor) else box.xywh
            x_center, y_center, width, height = bbox.flatten()

            # Write to file
            f.write(f'{class_id} {conf:.4f} {x_center:.2f} {y_center:.2f} {width:.2f} {height:.2f}\n')

    # Save the annotated image (optional)
    annotated_image = result.plot()
    cv2.imwrite(os.path.join(output_folder, image_file), annotated_image)  # Save in original BGR format
```
